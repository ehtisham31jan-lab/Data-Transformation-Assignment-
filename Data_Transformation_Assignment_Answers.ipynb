**Q1. Handle Missing Values (Basic Cleaning)**

Fill missing values using the following:

* Age → Median

* Salary → Mean

* Hours_Worked_Per_Week → Median

* Performance_Score → Mean

Display the dataset after handling missing values

**Q7. Build Preprocessing Pipeline**

Create a pipeline that:

* Applies encoding to categorical columns

 *Applies scaling to numerical columns

Use:

* ColumnTransformer

* Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler, MinMaxScaler

# Define columns
num_median_cols = ['Age', 'Hours_Worked_Per_Week']
num_mean_cols = ['Salary', 'Performance_Score']
num_no_impute = ['Projects_Completed']

label_encode_cols = ['Gender', 'Department']
onehot_cols = ['Work_Mode', 'Location']

# 1. Numerical Pipelines
num_median_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

num_mean_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
])

num_proj_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# 2. Categorical Pipelines
label_pipeline = Pipeline(steps=[
    ('ordinal', OrdinalEncoder())
])

onehot_pipeline = Pipeline(steps=[
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])

# 3. Combine in ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num_median', num_median_pipeline, num_median_cols),
    ('num_mean', num_mean_pipeline, num_mean_cols),
    ('num_proj', num_proj_pipeline, num_no_impute),
    ('label', label_pipeline, label_encode_cols),
    ('onehot', onehot_pipeline, onehot_cols)
], remainder='drop')

print("Preprocessing pipeline built successfully!")
```

**Q8. Apply Pipeline**

Apply the pipeline on the dataset.

Display:

* Transformed dataset

* Shape of final dataset

```python
# Apply the pipeline
transformed_data = preprocessor.fit_transform(df)

# For better display, let's convert the output back to a DataFrame (extracting column names)
ohe_feature_names = preprocessor.named_transformers_['onehot']['onehot'].get_feature_names_out(onehot_cols)
all_feature_names = num_median_cols + num_mean_cols + num_no_impute + label_encode_cols + list(ohe_feature_names)

final_df = pd.DataFrame(transformed_data, columns=all_feature_names)

print("Final Transformed Dataset:")
display(final_df.head())

print(f"\nShape of final dataset: {final_df.shape}")
```

**Q9. Conceptual Question**

Why is scaling important in Python?

**A9.** Scaling is critical in Machine Learning because many models (like K-Nearest Neighbors, Support Vector Machines, and Neural Networks) use distance calculations (like Euclidean distance) or gradient descent optimizations. If features are on entirely different scales (e.g., Age ranges from 20-60, but Salary ranges from 40,000-100,000), the model will heavily weigh the larger numbers (Salary) and effectively ignore the smaller ones (Age), leading to biased predictions and slower model training. Scaling brings all features to a level playing field.

**Q10. Conceptual Question**

Why do we convert categorical data into numerical form?

**Answer:** Machine learning models are fundamentally based on mathematical equations and statistical calculations. They cannot compute operations on text strings (like multiplying "Remote" by "Finance"). Therefore, converting categorical text into a numerical format (via Label Encoding or One-Hot Encoding) is necessary so the algorithms can read, process, and extract patterns from these features mathematically.
